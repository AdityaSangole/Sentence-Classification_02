```python

# Import necessary libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
# Load your labeled dataset
data = pd.read_excel('/content/immverse_ai_eval_dataset (2).xlsx')

# Split the dataset into training (60%), validation (20%), and test (20%) sets
train_data, temp_data =train_test_split(data, test_size=0.4,random_state=42)
valid_data, test_data = train_test_split(temp_data, test_size=0.5, random_state=42)
```

```python
data
```

```python

# Preprocessing: Convert text data into numerical features (e.g., TF-IDF)
tfidf_vectorizer =TfidfVectorizer(max_features=10000)
# You can adjust max_features
X_train = tfidf_vectorizer.fit_transform( train_data['sentence'])
X_valid = tfidf_vectorizer.transform(valid_data['sentence'])
X_test = tfidf_vectorizer.transform(test_data['sentence'])
```

```python

# Prepare target labels
y_train = train_data['voice']
y_valid = valid_data['voice']
y_test = test_data['voice']
```

```python
#Train a text classification model
model =RandomForestClassifier(random_state=42)

model.fit(X_train, y_train)
```

```python
# Evaluate the model on the validation set
y_valid_pred = model.predict(X_valid)
accuracy = accuracy_score (y_valid, y_valid_pred)
print(f"Validation Accuracy:{accuracy:.2f}")
```

```python
# Evaluate the model on the test set
y_test_pred = model.predict(X_test)
test_accuracy=accuracy_score (y_test,y_test_pred)
print(f"Test Accuracy: {test_accuracy:.2f}")
```

```python

# Generate a classification report for more detailed metrics
classification_rep =classification_report(y_test, y_test_pred)
print("Classification Report:\n", classification_rep)
```
